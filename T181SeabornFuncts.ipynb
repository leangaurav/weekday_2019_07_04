```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.DataFrame(np.random.randint(1,20,30).reshape(10,3), columns = list('ABC'))
print(df)
```

```python
sns.set()
plt.plot(df[0], df[1],'o')
plt.show()
```

```python
print(dir(sns))
```

```python
sns.lmplot('A','B', df)
plt.show()
```

```python
df1 = pd.DataFrame(np.arange(1,31).reshape(10,3), columns = list('ABC'))
print(df)
```

```python
sns.lmplot('A','B', df1, hue ='C')
plt.show()
```

```python
df2 = pd.DataFrame(np.arange(1,21).reshape(10,2), columns = list('AB'))
df2['C'] = pd.Series([1,1,0,0,1,1,1,0,0,0])
print(df2)

sns.lmplot('A','B', df2, hue ='C')
plt.show()
```

```python
n = np.random.randint(1,100,1000)
sns.distplot(n)
plt.show()
```

```python
n = np.random.randint(1,100,1000)
sns.distplot(n, kde = False, bins = 20)
plt.show()
```
