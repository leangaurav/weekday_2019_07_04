# Drop

```python
idx = np.where(df.Type == 'Emigrants' )
print(idx)

dd = df.drop(idx[0], axis = 0)
print(dd.head())
```

```python
print(dd)
```

```python
df_d = dd.drop(['Type', 'Coverage', 'REG', 'DEV'], axis =1)
print(df_d)
```

```python
print(dd)
```

```python
df_d.rename(columns={"OdName": "Country", "RegName": "Region"})
```

```python
print(df_d.columns)

new_col = [str(col) for col in df_d.columns]
print(new_col)

d = dict(zip(df_d.columns, new_col))
print(d)

df_d.rename(columns = d, inplace = True)
print(df_d)

```
