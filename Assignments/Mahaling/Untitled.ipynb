```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'Titanic_cleaned.csv')
print(df.head())
```

```python
%matplotlib qt
%matplotlib inline
```

```python
print(df.head())
print(df.columns)
print(df.index)
```

```python
plt.figure(figsize = (20,10))
plt.plot(df['age'], 'r--')
plt.xlabel('ID')
plt.xlabel('Age')
plt.title('Line Plot for Age of Passengers')
plt.legend()

plt.show()
```
