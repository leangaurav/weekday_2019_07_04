```python
df.drop(['Unnamed: 0', 'cabin', 'body'], inplace = True, axis = 1)
```

```python
df.drop(['boat'], inplace = True, axis = 1)
```

```python
df.fare.replace(r'\$', '', regex = True, inplace = True)
```

```python
df.fare = df.fare.astype(float)
```

```python
df.dtypes
```

```python
print(df.isna().any())
```

```python
df.age.fillna(df.age.mean(), inplace = True)
df.fare.fillna(df.fare.mean(), inplace = True)
```

```python
print(df.isna().any())
```

```python
df.drop(['home.dest'], inplace = True, axis = 1)
```

```python
r = np.where(df.embarked.isna()) # Index of rows where #NA is present for 'embarked'
print(r)

df.drop(index = r[0], inplace = True, axis = 0) # dropping rows in place
```

```python
df.to_csv('Titanic_cleaned.csv')
```
