```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
print(dir(plt))
```

```python
n = np.arange(1,31).reshape(10,3)
df = pd.DataFrame(n)
df[3] = df[2] * df[2]
print(df)
plt.plot(df[0], df[1])
plt.plot(df[0], df[3],'r*')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot')
plt.show()
```

```python
%matplotlib qt
%matplotlib inline
plt.figure(figsize = (15,5))
plt.plot(df[0], df[1])
plt.plot(df[0], df[3],'r*')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot')
plt.savefig('plot.png')
plt.show()
```

```python

plt.figure(figsize = (15,5))

plt.subplot(1,2,1)
plt.plot(df[0], df[1])

plt.subplot(1,2,2)
plt.plot(df[0], df[3],'r*')
plt.savefig('plot.png')
plt.show()
```

```python

plt.figure(figsize = (15,5))

plt.plot(df[0], df[1], label = 'First')
plt.plot(df[0], df[3], label = 'Second')
plt.yticks(range(0,1000,100))
plt.xticks([0,2,4,6,8])
plt.legend()
plt.text(8,100,'text')
plt.show()
```

```python
label = ['first','ab','cd','last']
plt.figure(figsize = (15,6))
h = [10,3,4,50]
plt.subplot(1,2,1)
plt.bar(label,h)

for i in range(len(label)):
    plt.text(label[i],h[i] + 2, label[i])
    

plt.subplot(1,2,2)
plt.pie(h, labels = label, labeldistance = 0.5, shadow = True) # try explode option

plt.show()
```
