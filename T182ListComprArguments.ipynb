```python
x = 1
def funct1():
    x = 10
    print("funct1", x)
    def funct2():
        nonlocal x
        x = 20
        print("funct2", x)
    funct2()
    print("funct1", x)

print("Main", x)
funct1()
print("Main", x)
```

```python
x = 10,20
print(x)
```

```python
def funct(x,y):
    print(x,y)
    
funct(10,20)

funct( *(10,20) ) # tuple unpacking
```

```python
def funct(*args): #variable length arguments
    print(args, type(args))
    
funct(10,20)
funct()
funct(10)
```

```python
def funct(x,y):
    print(x,y)
    
d = {'x' : 10, 'y' : '20'}
funct(1,2)
funct(x = 100, y = 200)
funct(y = 200, x = 100000)
#funct(y = 200, x = 100000, z = 111)
funct(**d)
```

```python
def funct(**kwargs):# keyword arguments; No positional args
    print(kwargs)
    
d = {'x' : 10, 'y' : '20'}
#funct(1,2)
funct(x = 100, y = 200)
funct(y = 200, x = 100000, abc = False)
funct(**d)
funct()
print(*d)

d1 = {'sep':'#', 'end':'ABC'}
print(**d1)
```

```python
def f(*args, a = 10, b = 20):
    print(args, a, b)
    
f(10,20,1000,2000, a= -10)
```

```python
def f(*args, **kwargs):
    print(args, kwargs)
    
f()
f(10)
f(10,20)
f(a = 10)
f(1,2, a = 10)
```

```python
def f(x,y , z = 100, *args, **kwargs):
    print(x,y, z,args, kwargs)
    
#f()
#f(10)
f(10,20)
#f(a = 10)
f(1,2, a = 10)
```

## List Comprehension

```python
r = []
for i in range(10):
    r.append(i)
    
print(r)
```

```python
a = list(range(10))
print(a)
```

```python
r = []
for i in range(10):
    r.append(i*i)
    
print(r)
```

```python
[ expr for var in seq ]
```

```python
a= [ i*i for i in range(10)  ]
print(a)
```

```python
a= [ 1 for i in range(10)  ]
print(a)
```

```python
import random
a= [ random.randint(10,20) for i in range(10)  ]
print(a)
```

```python
s = input("Enter a string:")

a = []
for c in s:
    a.append(c)
print(a)
print( [x  for x in s])
```

```python
s = input("Enter a string:")
a = [ (x, ord(x))  for x in s]
print( a )
```

```python
a = [ (chr(x), x) for x in range(65,91)]
print(a)
```

```python
a = [ (chr(x), x) for x in range(65,91)]
d= dict(a)
print(d)
```

```python
d = { chr(x): x for x in range(65,91)}
print(d)
```

```python
a =  [input('string: e') for i in range(1,5)]
print(a)
```

```python
[ expr for var in seq if <condition>]
```

```python
s = input("Enter string:")

r = [    c            for c in s         if c in 'aeiou']
print(r)
```

```python
import random

nums = [random.randint(1,20) for _ in range(10) ]

res = [ "Even" if n%2 == 0 else "odd"     for n in nums]
print(nums)
print(res)
```
